```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
plt.jet()
```

```python
import sklearn.datasets as datasets
```

```python
X, Y = datasets.make_blobs(centers=4, cluster_std=0.7, random_state=0)
fig, axes = plt.subplots()
axes.scatter(X[:,0],X[:,1])
```

```python
fig, axes = plt.subplots()
axes.scatter(X[:,0],X[:,1],c=Y)
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(4)
Y_hat = kmeans.fit(X).labels_
```

```python
fig,axes = plt.subplots(2,1,figsize=(4,8))
axes[0].scatter(X[:,0],X[:,1],c=Y_hat)
axes[1].scatter(X[:,0],X[:,1],c=Y)
```

```python
fig, axes = plt.subplots()
axes.scatter(X[:,0],X[:,1],c=Y_hat,alpha=0.5)
mu = kmeans.cluster_centers_
axes.scatter(mu[:,0],mu[:,1],s=100,c=np.unique(Y_hat))
```

<h1> k-means images </h1>


```python
from sklearn.datasets import fetch_mldata
from sklearn.cluster import KMeans
from sklearn.utils import shuffle
X_digits, _,_, Y_digits = fetch_mldata("MNIST Original").values() # fetch dataset from internet
X_digits, Y_digits = shuffle(X_digits,Y_digits) # shuffle dataset (which is ordered!)
X_digits = X_digits[-5000:]  
```

```python
plt.rc("image", cmap="binary") # use black/white palette for plotting
for i in xrange(10):
    plt.subplot(2,5,i+1)
    plt.imshow(X_digits[i].reshape(28,28))
    plt.xticks(())
    plt.yticks(())

    plt.tight_layout()
```

```python
kmeans = KMeans(20)
X_red = kmeans.fit(X_digits).cluster_centers_
```

```python
plt.rc("image", cmap="binary") # use black/white palette for plotting
for i in xrange(10):
    plt.subplot(2,5,i+1)
    plt.imshow(X_red[i].reshape(28,28))
    plt.xticks(())
    plt.yticks(())

    plt.tight_layout()
```
