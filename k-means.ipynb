```python
%matplotlib inline
import numpy
```

<h1> Generating data </h1>
<p>Data is generated for 200 patients with their response for 10 symptoms each.

```python
sym = numpy.random.uniform(0,5,size=(1000,10))
sym[:10,:]
```

```python
type(sym)
```

<h1> K-means clustering <h1>

```python
from sklearn.cluster import KMeans
kmeans = KMeans(12)

sym_k = kmeans.fit(sym).labels_ #assuming 12 clusters
```

```python
sym_k[:5]
```

<h1> visualizing the clusters </h1>
<p> matplotlib is used to visualize the clusters, 2d and 3d plots are drawn

```python
import matplotlib
import matplotlib.pyplot as plt
matplotlib.use('Agg')
```

```python
# 3 features are chosen to represent the 3 axis in the plots
# change the variables to choose different features
x = 0
y = 5
z = 9
```

<h2> cluster centroid of each disease </h2>

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

centroid = kmeans.cluster_centers_

ax.scatter(centroid[:,x],centroid[:,y],centroid[:,z] ,marker='o', linewidths=10)


plt.show()
```

<h2> individual disease cluster <h2>

```python
cluster_num = int(raw_input("choose disease from 0 to 11"))
cn = cluster_num
```

```python
sym_clus = sym[sym_k == cluster_num]
sym_clus[:10]
```

```python

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(sym_clus[:,x], sym_clus[:,y], sym_clus[:,z], c=[1,0,0])
ax.scatter(centroid[cn,x],centroid[cn,y],centroid[cn,z] ,marker='o', linewidths=10)

plt.show()
```
